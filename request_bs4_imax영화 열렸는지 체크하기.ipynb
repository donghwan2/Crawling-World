```python
import requests
from bs4 import BeautifulSoup

import re
```

# 1. 상영날짜의 영화 제목 추출하기

## 1.1 방법1-select 함수와 정규표현식으로 영화제목 추출

```python
url = "http://www.cgv.co.kr/common/showtimes/iframeTheater.aspx?areacode=01&theatercode=0013&date=20220104&screencodes=&screenratingcode=&regioncode="
html = requests.get(url)
print(html.text)
```

```python
# html에 BeautifulSoup 모듈 적용
soup = BeautifulSoup(html.text, 'html.parser')
soup
```

```python
title_list = soup.select('.info-movie > a > strong')

print(len(title_list))
title_list
```

```python
# 리스트의 첫번째 데이터를 확인해보면, 텍스트는 맞는데 잘 모르는 데이터 타입.
i = 0
text = title_list[i]

print(type(text), '\n')
print(text)
```

```python
# 우리가 잘 아는 str 타입으로 변환해주고,
text = str(text)
text
```

```python
# 정규표현식을 사용해서 제목만 추출합니다.
regex = '[가-힣 ]+'

text_reg = re.findall(regex, text)[0]
text_ref = text_reg.strip()
text_ref
```

```python
# for문으로 묶어서 영화 전체에 대한 제목 추출하기
for i in range(len(title_list)):
    text = soup.select('.info-movie > a > strong')[i]
    text = str(text)
    regex = '[가-힣 -]+'  # 영화 제목에 '-'가 들어있는 것들이 있어서 추가.
    text_reg = re.findall(regex, text)[0]
    text_ref = text_reg.strip()
    print(text_ref)
```

## 방법2. select 함수와 text 함수로 영화제목 추출

```python
# select_one 함수를 이용하면 원하는 데이터 중 첫번째를 추출할 수 있습니다.
title_first = soup.select_one('div.info-movie > a > strong')
title_first
```

```python
# select 함수를 이용하면 원하는 데이터 전체를 리스트로 추출할 수 있습니다.
title_list = soup.select('div.info-movie > a > strong')
title_list
```

```python
# 첫번째 영화 제목
print(type(title_list[0]))
print(title_list[0])
print(title_list[0].text.strip())  # 텍스트 추출 후 공백 제거
```

```python
# for문으로 리스트 전체 데이터에 적용하기
for i in title_list:
    print(i.text.strip())
```

# 2. IMAX가 열렸으면 해당 영화제목 추출하기

```python
imax = soup.select_one('.imax')
imax
```

```python
if imax:
    print('IMAX 예매가 열렸습니다.')
else:
    print('IMAX 예매가 열리지 않았습니다.')
    
```

```python
# 만약 imax가 있으면 부모 class 중 'col-times'(영화 블록)를 찾습니다.
# 그리고 그 밑에 영화제목 정보를 가져옵니다.
if imax:
    imax = imax.find_parent('div', class_='col-times')
    title = imax.select_one('div.info-movie > a > strong').text.strip()
    print(title + " IMAX 예매가 열렸습니다.")
else:
    print('IMAX 예매가 열리지 않았습니다.')
```
